```python
import numpy as np
def relu(x): 
    return np.maximum(x, 0.0)
def relu_derivative(x):
    return 1
def smell(in_num,l_num,k_num,out_num,weights,x,y):#权重矩阵，输入和输出的训练集合
    y_y=x_y=len(x)
    #print(len(x))
    x_x=len(x[0])
    y_out=0
    #划分权重区间
    w_xw=[]
    
    for q in range(in_num):
        w_x=[]
        for p in range(k_num):
            w_x.append(weights[(q+1)*(p+1)-1])
        w_xw.append(w_x)
    
    b_xw=[]
    for b in range(k_num):
        b_xw.append(weights[30+b])
    
    w_wy=[]
    for pp in range(k_num):
        w_wy.append(weights[40+pp])
    b_wy=weights[-1]    
    #前向计算
    #print(x)
    x=np.array(x)
    w_xw=np.array(w_xw)
    b_xw=np.array(b_xw)
    w_wy=np.array(w_wy)
    y_result=[]
    for t in range(x_y):
        y_result.append(relu(np.dot(relu(np.dot(x[t],w_xw)+b_xw),w_wy)+b_wy))
    y_error=[]
    for tt in range(x_y):
        y_error.append(abs(y[tt]-y_result[tt]))
    y_error_v=0
    y_error_sum=0
    for ttt in range(x_y):
        y_error_sum=y_error_sum+y_error[ttt]
    y_error_v=y_error_sum/x_y
    #print('b_xw',b_xw)
    #print('y_result',y_result)
    ##print("x",x[0])
    #print(b_xw)
    ##print("b_wy",w_wy)
    #print("b_yw",b_wy)
    #print(weights)
    #print('jiesu')
    #print("结果",y_result) 
    #print("误差",y_error)
    #print('平均误差',y_error_v)
    return y_error_v
```

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Oct 23 18:11:35 2018

@author: wangsi
"""
import random
def foa(in_num,l_num,k_num,out_num):#输入，隐含层数，隐含层粒子数，输出个数
    # 随机初始果蝇群体位置 
    x_init=0.5*random.random() 
    y_init=0.5*random.random() 
    #print(x_init,y_init) 
    maxgen=2000 # 迭代次数 
    size=10#种群规模
    sizepop=0 # 粒子个数
    sizepop=sizepop+in_num*k_num+out_num+k_num*out_num
    for i in range(l_num):
        sizepop=k_num*k_num+sizepop+k_num
    sizepop=sizepop-k_num*k_num
    #print(sizepop)
    x=[[0]*sizepop for i in range(size)]#横为sizepop,z纵为sizex[1][sizepop]
    y=[[0]*sizepop for i in range(size)] 
    D=[[0]*sizepop for i in range(size)]
    S=[[0]*sizepop for i in range(size)]
    Smell=[0]*size
    
    yy=[0]*maxgen #每次的最优值
    w_best=[]#每次最优权重系数
    X_best=[[0]*sizepop for i in range(maxgen)]#X_best[maxgen][sizepop]
    Y_best=[[0]*sizepop for i in range(maxgen)]
    # 果蝇寻优开始，利用嗅觉寻找食物 
    for i in range(size):
        for j in range(sizepop):
            x[i][j]=x_init+2*random.random()-3 
            y[i][j]=y_init+2*random.random()-3
            D[i][j]=(x[i][j]**2+y[i][j]**2)**(0.5) 
            S[i][j]=1/D[i][j]
        Smell[i]=smell(in_num,l_num,k_num,out_num,S[i],X_train,y_train)
    #print(Smell)
    # 找出此果蝇群体的中味道浓度最高的果蝇 ( 求极大值 ) 
    bestSmell=min(Smell) 
    bestindex=Smell.index(bestSmell) 
    #print(bestindex)
    # 保留最佳味道浓度值与 x 、 y 坐标，此时果蝇群体利用视觉往该位置飞去。 
    x_init=x[bestindex];y_init=y[bestindex]; 
    Smellbest=bestSmell 
    #print("xzuiyouzhi",x_init)
    # 果蝇迭代寻优开始 
    learn=0.001
    for g in range(maxgen): 
        # 附与果蝇个体利用嗅觉搜寻食物之随机方向与距离
        for i in range(size):
            for j in range(sizepop):
                x[i][j]=x_init[j]+(2*random.random()-1)*learn
                y[i][j]=y_init[j]+(2*random.random()-1)*learn
                D[i][j]=(x[i][j]**2+y[i][j]**2)**(0.5) 
                S[i][j]=1/D[i][j] 
            # 判定值 (S) 代入判定函数以求出该果蝇位置的味道浓度 (Smelli) 
            Smell[i]=smell(in_num,l_num,k_num,out_num,S[i],X_train,y_train)
        # 找出此果蝇群体的中味道浓度最高的果蝇 ( 求极大值 ) 
        bestSmell=min(Smell) 
        bestindex=Smell.index(bestSmell) 
        # 判断味道浓度是否优于前一迭代味道浓度， 
        # 若是则保留最佳味道浓度值与 x 、 y 坐标，此时果蝇群体利用视觉往该位置飞去。 
        if bestSmell<Smellbest: 
            x_init=x[bestindex];y_init=y[bestindex]
            Smellbest=bestSmell
            w_best.append(S[bestindex])
        # 每代最优 Smell 值纪录到 yy 数组中，并记录最优迭代坐标 
        yy[g]=Smellbest 
        X_best[g]=x_init;Y_best[g]=y_init 
        print(Smellbest)
    #for i in range(100):
        #print('x'+str(X_best[i]))
        #print('y'+str(Y_best[i]))
        #print(yy[i])
    #print("") 
    #print ("经过果蝇算法优化后的最优结果：") 
    #print (Smellbest)
    return w_best
```
